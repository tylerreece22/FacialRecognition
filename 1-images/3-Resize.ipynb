```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image = cv2.imread('../test.jpg') # RGB format
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image, cmap='gray')
```

```python
# shrink
image_re = cv2.resize(gray_image, (500,700), cv2.INTER_AREA)
image_re.shape
```

```python
plt.imshow(image_re,cmap='gray')
```

```python
# enlarge
image_re = cv2.resize(gray_image, (2000,3000), cv2.INTER_CUBIC)
image_re.shape
```

```python
plt.imshow(image_re,cmap='gray')
```
