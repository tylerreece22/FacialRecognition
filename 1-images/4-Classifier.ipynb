```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
image = cv2.imread('../data/me.jpg')
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray,cmap='gray')
```

```python
haar = cv2.CascadeClassifier('../data/haarcascade_frontalface_default.xml')
faces = haar.detectMultiScale(gray, 1.3,5)
```

```python
print(faces)
```

```python
# show rectangle on detected face
top_left = faces[0][0]
bottom_left = faces[0][1]
length = faces[0][3]
cv2.rectangle(image, (top_left,bottom_left), (top_left+length,bottom_left+length), (0,255,0),10)
cv2.imwrite('../data/me_full_size.png', image)
plt.imshow(image)
```
